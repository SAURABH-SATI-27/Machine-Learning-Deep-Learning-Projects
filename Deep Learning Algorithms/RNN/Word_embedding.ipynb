```
from tensorflow.keras.preprocessing.text import one_hot
```

```
sent = ['I am a Machine Learning Engineer' , 'We are learing Word Embedding Technique' ,'LSTM is more powerful for Natural Language Processing','Data is love' ]
```

```
voc_size = 1000 # size of vocabulary dict
```

```
# One_hot_Encoding
one_hot_rep = [one_hot(word , voc_size)for word in sent]
```

```
one_hot_rep  # we will get index of every letter from our dictionary 
```

```
sent_len = 10 # it must be equal or greater than max sent len
```

```
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
import pandas as pd
import numpy as np
```

```
embedded_docs = pad_sequences(one_hot_rep , padding = 'pre', maxlen = sent_len)
```

```
dim =10
model=Sequential()
model.add(Embedding(voc_size  , dim , input_length=sent_len))
```

```
model.compile('adam' , 'mse')
```

```
model.summary()
```

```
model.predict(embedded_docs) # now this will be the vector representation features of every sentence with 10 dimensions for each words
```
